```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x-axis (days of the week) for the sinusoidal graph
x_days = np.linspace(0, 4, 500)  # Smooth values from Monday (0) to Friday (4)
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]

# Adjust the frequency of the sinusoidal curve to 5 times a day
y_sin_mid_freq = 85 + 15 * np.sin(10 * np.pi * x_days - np.pi / 2)  # 5 oscillations per day

# Plot the medium-frequency sinusoidal graph
plt.figure(figsize=(8, 5))
plt.plot(x_days, y_sin_mid_freq, label="Resetting to 100 with Salawat and Istighfar", color='purple', linewidth=2)

# Customize the axes
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Add horizontal line at y=0
plt.xticks(ticks=np.arange(5), labels=days)
plt.yticks(np.arange(0, 101, 10))  # Extend y-axis from 0 to 100
plt.ylim(0, 100)

# Add labels, title, and legend
plt.xlabel("Days of the Week")
plt.ylabel("Spiritual Level (%)")
plt.title("Resetting to 100 with Salawat and Istighfar")
plt.legend()
plt.grid(alpha=0.5)
plt.tight_layout()

# Show the plot
plt.show()
```
